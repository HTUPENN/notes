```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
```

```python
data = pd.read_csv("./课件/07数据分析篇/smoke.csv")
```

```python
data
```

1. 分类 -> 虚拟变量

```python
logistic_data = pd.get_dummies(data, columns=["gender", "smoking"],
                               dtype=int, 
                               drop_first=True)
logistic_data
```

2. 相关性

```python
logistic_data.corr().abs() > 0.8
```

3. 创建模型

```python
y = logistic_data['cancer']
X = logistic_data.drop('cancer', axis=1)

# 在自变量里添加一个常量（为了引入截距）
X = sm.add_constant(X)
```

4. 训练模型

```python
model = sm.Logit(y, X).fit()
model.summary()
```
